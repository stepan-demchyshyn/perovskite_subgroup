## 0. Motivation and technical notes

This short guide is written in order to assist with scientific figure preparation for bachelor/master/PhD thesis or publication in a scientific journal focusing on using Jupyter and Python. **It is not meant to be the definitive set of rules**, but a starting point from which to begin your figure drafts.

The guide **assumes some basic knowledge of python** (including importing data, basic data processing and basic knowledge of matplotlib as a plotting tool). It will mostly focus on figure preparation, keeping in mind both clarity of presentation and Ã¦sthetics.  

Here python [**matplotlib**](https://matplotlib.org/stable/gallery/index.html) is chosen as the main tool, though other libraries exist that can produce comparable results ([seaborn](https://seaborn.pydata.org/), [plotly](https://plotly.com/), ggplot, *etc*.). 

In order to easily arrange your plots into great looking figures use [**Inkscape**](https://inkscape.org/) vector graphics software (**FREE**). An Inkscape template is also included in [the github page](https://github.com/stepan-demchyshyn/perovskite_subgroup) in a separate folder. Alternatives to Inkscape include CorelDraw, Adobe Illustrator, PowerPoint, etc. Skills learned in Inkscape are transfarable to other software and frees you up from expensive licenses. 

Always first consider stylistic requirements put in place by JKU (or your group, scientific journal, etc.) and then refer to this guide. 

This guide is partialy based on [Science Magazine figure prep guide](https://www.sciencemag.org/sites/default/files/Figure_prep_guide.pdf). 

## 1. General considerations

### Be aware of the format and size. 

One of the first steps in preparing graphics is to decide on the format and size of your final figure and where it will be displayed (printed media [thesis, papers, etc.], slide presentations [power point, etc.], or poster). While seemingly trivial, the way your figurues and plots appear on the computer screen quite often do not translate directly to other types of media. Being aware of the size and format will prevent you from accidently making figures that are hard to read *i.e.* fonts too small, lines or markers too thin, *etc*.

Most of the printed media will be set to fit on the A4 format, except of the publications located in North America, where the standard is US-letter. The differences between them are minute and a general figure size described below should fit both of the formats well. 

### Exporting and saving files

When exporting your final figure the best practice is to **use vector formats** (svg, pdf, etc.) or image files with resolution no less than 300 dpi (png, jpeg, tiff). 

**HINT**: When saving individual panels and bringing them into Inkscape, vector files will allow you to do some final adjustments in there while arranging your panels. This is very useful for minor changes required at the very end of thesis/paper writing, *e.g.* when you spot a typo or need to change a color of a line (easier to make changes in Inkscape instead of remaking the whole panel/figure). 

## 2. Sizing of panels

Borrowed from [Science Magazine Figure Prep Guide](https://www.sciencemag.org/sites/default/files/Figure_prep_guide.pdf) the outline below aims for the whole figure to fit within 184 mm wide box. The rest of the individual panel sizes are calculated based on this width limit  (assuming they will occupy either full page, half, third or a quarter of the page [fifth also works for some panels]) giving 4 mm space in between them. The width of the graphs is calculated in order to nicely fit everything together. The height is mostly a suggestion and can be modified without causing too much disruption.

In the table below you fill find the dimensions of the various panels calculated. 

| Object           | Width (mm)    | Height (mm)                        |
| -----------------|--------------:|-----------------------------------:|
| Full page (A4)   | 210           | 297                                |
| Figure           | 184           |                                    |
| Full panel       | 184           | <span style="color:gray">45</span> |
| Half-panel       | 90            | <span style="color:gray">55</span> |
| Third-panel      | 58.67         | <span style="color:gray">55</span> |
| Quarter-panel    | 43            | <span style="color:gray">45</span> |

## 3. Panel style

Taking into account overall size of printed page the following style can be used: 

* Font: Arial 
* Axes linewidth min. 0.8 pt
* Y and X axes label 9 pt (units in parentheses)
* Tick labels 7 pt
* Major ticks facing in on all sides, length 2.5 pt
* Line plot: min 1.25 pt
* Scatter plot marker size: min. 4 pt
* Legend without a frame, min. font size 5 pt

All the aforementioned things are included in this .mplstyle sheet. By using it you will automatically apply the changes to all plots in this jupyter notebook. 

```python
plt.style.use('./publication.mplstyle')
```

Selection of good color scheme for panels and figures is a way to elevate the quality of your publication. There are several considerations that need to be taken into account when selecting a color scheme. An ideal color scheme should be:

* Inclusive
* Media independent
* Aestheticly appealing

Suggested color scheme below is inspired by this and this plots from The Economist. Its distinguishable for person with protanopia , deuteranopia  and tritanopia types of dichromacy color blindness. It uses bright colors which should be well distinguishable on different types of monitors and prints.

```python
cmap = (['#ff5555', '#2ec1d3', '#0083c5', '#ffb945', '#7e7e7e'])
```

```python
figsize = (3.0, 1.0) #1/1 
fig, axs = plt.subplots(figsize=figsize)

cmap = (['#ff5555', '#2ec1d3', '#0083c5', '#ffb945', '#7e7e7e'])
x = 0
y = 1

for i, color in enumerate(cmap):
    xy = np.array([x, y])
    axs.add_patch(plt.Circle(xy, radius=0.5, color=color))
    x = x + 1 
axs.axis('equal')
axs.set_xlabel(r'Color number')

labels = np.arange(0, 5, 1)

axs.set_xlim(0, 4)

axs.set_xticks(labels)

axs.tick_params(axis = 'y',labelleft = False, length = 0 )

plt.tight_layout()
```

### Histograms

```python
data = pd.read_csv('./data/stats_example.dat')

data.head()
```

```python
from scipy.stats import norm
import matplotlib.mlab as mlab
from scipy.optimize import curve_fit

def gauss(x, H, A, x0, sigma):
    return H + A * np.exp(-(x - x0) ** 2 / (2 * sigma ** 2))

def gauss_fit(x, y):
    mean = sum(x * y) / sum(y)
    sigma = np.sqrt(sum(y * (x - mean) ** 2) / sum(y))
    popt, pcov = curve_fit(gauss, x, y, p0=[min(y), max(y), mean, sigma])
    return popt
```

```python
figsize = (2.310, 2.165) #1/3 
fig, axs = plt.subplots(figsize=figsize)

n, bins, patches = axs.hist(data['PCE'], bins = 10,  rwidth=0.9, color = cmap[2], alpha = 1)

#estimating a pdf 
(mu, sigma) = norm.fit(data['PCE'])
x = np.arange(0, 22, 0.1)
y = norm.pdf(x, mu, sigma)
axs.plot(x, y*270, color = cmap[0]) #

axs.set_ylim(0, 74)
axs.set_xlim(9.5, 21.5)

axs.set_ylabel('Number of devices')
axs.set_xlabel('Effciency (%)')

axs.set_xticks(np.arange(10, 22, 2))

plt.tight_layout()

#plt.savefig('./plots/histogram_example.svg')
```

### Box plot

```python
figsize = (2.310, 2.165) #1/3 
fig, axs = plt.subplots(figsize=figsize)
num_col = 4 
data = []
for i in range(num_col):
    r = np.random.uniform(low=-1, high=1, size=(1,))
    mu, sigma = (i+r[0])*0.25, 0.5
    subset = np.random.normal(mu, sigma, 100) 
    data.append(subset)
bp = axs.boxplot(data, patch_artist = True)

#plt.setp(bp["boxes"], facecolor=cmap[1]) #change all colors together

color = ([cmap[0], cmap[1], cmap[2], cmap[3], cmap[4]])
for patch, color in zip(bp['boxes'], color):
    patch.set_facecolor(color)

axs.set_ylabel(r'Y axis (units)')
axs.set_xlabel(r'X axis (units)')

plt.tight_layout()

#plt.savefig('./plots/boxplot_example.svg')
```

```python
figsize = (2.310, 2.165) #1/3 
fig, axs = plt.subplots(figsize=figsize)

delta = 0.025
x = y = np.arange(-3.0, 3.0, delta)
X, Y = np.meshgrid(x, y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = (Z1 - Z2) * 2

im = axs.imshow(Z, interpolation='bilinear', cmap = 'inferno',
               origin='lower', extent=[-3, 3, -3, 3],
               vmax=abs(Z).max(), vmin=-abs(Z).max())


axs.set_ylabel(r'Y axis (units)')
axs.set_xlabel(r'X axis (units)')

plt.tight_layout()

#plt.savefig('./plots/heatmap_example.svg')
```
